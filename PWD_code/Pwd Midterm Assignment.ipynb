Globally, fertility is on the decline. As a result, in the absence of immigration, this region is set to experience the most rapid population aging and decline[1].

In the case of Singapore, the government has grappled with the relentless downward trend in fertility since the 1980s. -Through the exploratory data analysis from the year 2009-2018, we have been able to see that the value of the total fertility rate is gradually declining from the year 2009 - 2018. There is a sharp increase in the value from the year 2011 to 2012 as 2012 is the year of the dragon where most of the chinese families want to have their kid to be borned that year. After 2014, the value of the fertility rate is gradually dropping. 

Next, as we can witness from the bar chart or line chart, the majority of the female would like to have kids the age of 25 to 39 years old. These three groups 25-29, 30-34, 35-39 contribute nearly 90% of the fertility rate in Singapore. The percentage of these age groups are 28.76%,28.91%, and 19.26% respectively. In addition, the chart also indictate that most females would like to have children at the age of 30-34 years old. 

Furthermore, from the percentage change of fertility rate from 2009 - 2018, the total fertility rate for 15-19, 20-24, 25-29,30-34 are all dropping in 2009-2018. But, there is a growth in the percentage of 35-39, and 40-44, this might tell us that more females would like to have children at their later stage which this is might one of the reasons why the total fertility rate is gradually dropping in the past few years.
